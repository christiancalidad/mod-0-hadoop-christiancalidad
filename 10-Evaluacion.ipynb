## Mapper

```python
%%writefile mapper.py
#! /usr/bin/env python

##
## Esta es la función que mapea la entrada a parejas (clave, valor)
##
import sys
if __name__ == "__main__": 
    
    ##
    ## itera sobre cada linea de código recibida
    ## a través del flujo de entrada
    ##
    for line in sys.stdin:
        
        ##
        ## genera las tuplas palabra \tabulador 1
        ## ya que es un conteo de palabras
        ##
        word=line.split()[0:2]
        valor=line.split()[2]
        valor = int(valor)   

                   
            ##
            ## escribe al flujo estándar de salida
            ##
        sys.stdout.write("{}\t{}\n".format(word, valor))
```

## Reducer

```python
%%writefile reducer.py
#!/usr/bin/env python

import sys

##
## Esta función reduce los elementos que 
## tienen la misma clave
##

if __name__ == '__main__': 
  
    vector = []
    
    ##
    ## cada linea de texto recibida es una 
    ## entrada clave \tabulador valor
    ##
    for line in sys.stdin:
        
        key, val = line.split("\t") 
        val = int(val)

        if len(vector) < 6:
            vector.append((key, val))
        else:
            vector = sorted(vector, key=lambda tup: tup[1])
            if val < vector[5][1]:
                vector.pop(5)
                vector.append((key, val))

                    

for i in vector:            
    sys.stdout.write("{}\t{}\t{}\n".format(i[0][2:3], i[0][7:17], i[1])) 
               
```
